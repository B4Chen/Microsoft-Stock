```python
import pandas as pd
import matplotlib.pyplot as plt 

df = pd.read_csv("/Users/brian/Pandas/Microsoft_Stock/Microsoft Stocks.csv")
df
```

```python
df.set_index('Date')
```

```python
# Convert 'Date' column to datetime format
df['Date'] = pd.to_datetime(df['Date'])
plt.plot(df['Date'], df['Price'])
plt.ylabel('Price')
plt.xlabel('Date')
plt.xticks(rotation=45)  
plt.show()

```

```python
# Calculate the percentage change in stock price between consecutive years
df['hPrice'] = df['Open'].pct_change() * 100

# Find the year with the highest percentage change
highest_growth_year = df.loc[df['hPrice'].idxmax(), 'Date']

print("The highest growth year is:", highest_growth_year)
```

```python
# Assuming the 'Date' column is in datetime format
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

average_price_by_month = df.groupby('Month')['Price'].mean()
average_price_by_year = df.groupby('Year')['Price'].mean()

print("Average price by month:")
print(average_price_by_month)

print("Average price by year:")
print(average_price_by_year)
```

```python
# Assuming the 'Date' column is in datetime format
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month

volume_comparison_by_month = df.groupby('Month')['Volume'].sum()
volume_comparison_by_year = df.groupby('Year')['Volume'].sum()

print("Volume comparison by month:")
print(volume_comparison_by_month)

print("Volume comparison by year:")
print(volume_comparison_by_year)
```

```python
import matplotlib.pyplot as plt

# Assuming you have already calculated the average price by year
average_price_by_year = df.groupby('Year')['Price'].mean().tail(10)

# Plotting the average price by year
plt.plot(average_price_by_year.index, average_price_by_year.values)
plt.xlabel('Year')
plt.ylabel('Average Price')
plt.title('Average Price by Year (Last 10 Years)')
plt.show()
```
